Oppgave 1)
Lag et kurveplot (linjediagram) som viser utviklingen av temperaturen over en uke ut fra
følgende målinger:
dager = ['Mandag', 'Tirsdag', 'Onsdag', 'Torsdag', 'Fredag', 'Lørdag', 'Søndag']
temperaturer = [20, 22, 19, 24, 25, 23, 21]




```python
import numpy as np
import matplotlib.pyplot as plt


dager = ["Mandag", "Tirsdag", "Onsdag", "Torsdag", "Fredag", "Lørdag", "Søndag"]
temperaturer = [20, 22, 19, 24, 25, 23, 21]

plt.plot(dager, temperaturer)
plt.show()
```

Oppgave 2)
Lag et stolpediagram som viser antall bøker lest av forskjellige personer, gitt
informasjonen:
navn = ['Alice', 'Bob', 'Charlie', 'David']
antall_boker = [5, 3, 8, 2]



```python
import matplotlib.pyplot as plt

navn = ["Alice", "Bob", "Charlie", "David"]
antall_boker = [5, 3, 8, 2]
plt.bar(navn, antall_boker)
plt.show()
```

Oppgave 3)
Lag et sektordiagram som viser fordelingen av frukter ut fra følgende informasjon:
frukter = ['Eple', 'Banan', 'Appelsin', 'Druer']
andeler = [30, 25, 20, 25]



```python
import matplotlib.pyplot as plt

frukter = ['Eple', 'Banan', 'Appelsin', 'Druer']
andeler = [30, 25, 20, 25]

plt.pie(andeler, labels=frukter, autopct='%1.1f%%')
plt.show()
```

Oppgave 4)
Lag et scatterplot som viser sammenhengen mellom høyde og vekt for følgende data:
hoyde = [184, 162, 159, 188, 190, 165, 161, 181, 150, 179, 190]
vekt = [77, 81, 63, 67, 93, 65, 54, 78, 57, 61, 86]



```python
import matplotlib.pyplot as plt

hoyde = [184, 162, 159, 188, 190, 165, 161, 181, 150, 179, 190]
vekt = [77, 81, 63, 67, 93, 65, 54, 78, 57, 61, 86]

plt.scatter(hoyde, vekt)
plt.title("Høyde versus vekt")
plt.xlabel('Hoyde')
plt.ylabel('Vekt')
plt.show()
```
